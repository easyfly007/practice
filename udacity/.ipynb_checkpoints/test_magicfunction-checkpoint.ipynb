```python
from math import sqrt
def fib1(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fib1(n-1)+ fib1(n-2)

def fib2(n):
    return ((1+sqrt(5))**n - (1-sqrt(5))**n)/(2**n*sqrt(5))
```

```python
%timeit fib1(20)
```

```python
%timeit fib2(20)
```
